```python
#Libraries
import pandas as pd
import requests
import numpy as np
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd
import plotly.express as px


from io import StringIO
pd.set_option('display.max_columns', None)


```

```python
proximos_jogos_url="http://api.clubelo.com/Fixtures"

response=requests.get(proximos_jogos_url)

data=StringIO(response.text)
df=pd.read_csv(data)
df.head()
```

```python
df.columns
```

```python
df['Casa_Ganha']=df[['GD=1', 'GD=2', 'GD=3', 'GD=4', 'GD=5','GD>5']].sum(axis=1)
df['Empate']=df['GD=0']
df['fora_Ganha']=df[['GD<-5', 'GD=-5', 'GD=-4', 'GD=-3','GD=-2', 'GD=-1']].sum(axis=1)

#odd=1/probabilidade
# Calculando as probabilidades (odds)
df['Casa_odd'] = np.where(df['Casa_Ganha'] != 0, round(1 / df['Casa_Ganha'], 2), 0)
df['Empate_odd'] = np.where(df['Empate'] != 0, round(1 / df['Empate'], 2), 0)
df['fora_odd'] = np.where(df['fora_Ganha'] != 0, round(1 / df['fora_Ganha'], 2), 0)


df.head()
```

```python
df.columns
```

```python
df_final=df[['Date', 'Country', 'Home', 'Away','Casa_Ganha','Empate','fora_Ganha', 'Casa_odd', 'Empate_odd', 'fora_odd']]

df_final.head()
#Aqui eu renomiei algumas colunas
df_final.rename(columns={'Country': 'Campeonato', 'Home': 'Time Casa', 'Away': 'Time Visitante'}, inplace=True)


# Calcular a probabilidade de vitria do time da casa em porcentagem
df_final.loc['Probabilidade_Vitoria_Casa'] = round((1 / df_final['Casa_odd']) * 100, 2).astype(str) + '%'

# Calcular a probabilidade de derrota do time da casa em porcentagem
df_final.loc['Probabilidade_Derrota_Casa'] = round((1 / df_final['fora_odd']) * 100, 2).astype(str) + '%'

# Calcular a probabilidade de empate em porcentagem
df_final.loc['Probabilidade_Empate'] = round((1 / df_final['Empate_odd']) * 100, 2).astype(str) + '%'

# Calcular a probabilidade de vitria do time visitante em porcentagem
df_final.loc['Probabilidade_Vitoria_Fora'] = round((1 / df_final['fora_odd']) * 100, 2).astype(str) + '%'

# Calcular a probabilidade de derrota do time visitante em porcentagem
df_final.loc['Probabilidade_Derrota_Fora'] = round((1 / df_final['Casa_odd']) * 100, 2).astype(str) + '%'

df_final.reset_index(drop=True, inplace=True)
############

app = dash.Dash(__name__)

# Grfico de barras para probabilidade de vitria do time da casa
fig_vitoria_casa = px.bar(df_final, x='Time Casa', y='Casa_odd', title='Probabilidade de Vitria do Time da Casa',
                           labels={'Casa_odd': 'Probabilidade (%)', 'Time Casa': 'Time da Casa'})

# Grfico de barras para probabilidade de vitria do time visitante
fig_vitoria_fora = px.bar(df_final, x='Time Visitante', y='fora_odd', title='Probabilidade de Vitria do Time Visitante',
                           labels={'fora_odd': 'Probabilidade (%)', 'Time Visitante': 'Time Visitante'})
fig_empate = px.bar(df_final, x=df_final.index, y='Empate_odd', title='Probabilidade de Empate',
                    labels={'Empate_odd': 'Probabilidade (%)', 'index': 'Jogo'})


app.layout = html.Div([
    html.H1("Dados de Apostas"),
    dcc.Graph(id='fig_vitoria_casa', figure=fig_vitoria_casa),
    dcc.Graph(id='fig_vitoria_fora', figure=fig_vitoria_fora),
    dcc.Graph(id='fig_empate', figure=fig_empate)
])

if __name__ == '__main__':
    app.run_server(debug=True)




```
